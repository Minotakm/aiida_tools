```python
from aiida import orm, load_profile


load_profile()
```

```python
qb = orm.QueryBuilder()
qb.append(
    orm.Group, project=["label", "type_string", "*"], filters={"type_string": "core"}
)

groups = qb.all()
groups = [
    [label, type_string, node.nodes.__len__()] for label, type_string, node in groups
]
```

```python
def get_groups():
    qb = orm.QueryBuilder()
    qb.append(
        orm.Group,
        project=["label", "type_string", "*"],
        filters={"type_string": {"like": "core%"}},
    )
    rows = qb.all()
    return [
        {"label": label, "type_string": type_string, "n_nodes": len(group.nodes)}
        for label, type_string, group in rows
    ]
```
